# Question 1

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def calculate_I(k, alpha):
    normalization_constant = 1 / (2 * math.pi)
    exponential_term = np.exp(-k**2 / (4 * alpha))
    result = normalization_constant * np.sqrt(np.pi / alpha) * exponential_term
    return result

# Given alpha values
alpha_values = [0.1, 0.01, 0.001]

# Generate k values from -1 to 1
k_values = np.linspace(-1, 1, 1000)

# Plotting
plt.figure(figsize=(12, 8))

for alpha in alpha_values:
    I_values = [calculate_I(k, alpha) for k in k_values]
    plt.plot(k_values, I_values, label=f'α = {alpha}')

plt.xlabel('K')
plt.ylabel('I(k, α)')
plt.title('Plot of I(k, α) against K for specific α values')
plt.legend()
plt.grid(True)
plt.show()
```

# Question 2

```python
def calculate_L(k, alpha):
    return alpha / (np.pi * (k**2 + alpha**2))

# Given alpha values
alpha_values = [0.1, 0.01, 0.001]

# Generate k values from -1 to 1
k_values = np.linspace(-1, 1, 1000)

# Plotting
plt.figure(figsize=(12, 8))

for alpha in alpha_values:
    L_values = [calculate_L(k, alpha) for k in k_values]
    plt.plot(k_values, L_values, label=f'α = {alpha}')

plt.xlabel('K')
plt.ylabel('L(k, α)')
plt.title('Plot of L(k, α) against K for specific α values')
plt.legend()
plt.grid(True)
plt.show()
```
